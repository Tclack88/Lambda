## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}
2 & 4 & 6 \\
\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
A = np.array([[1,2],[3,4],[5,6]])
B = np.array([[2,4,6]])
C = np.array([[9,6,3],[4,7,11]])
D = np.array([[1,0,0],[0,1,0,],[0,0,1]])
E = np.array([[1,3],[5,7]])

combos = {'A':A.shape,'B':B.shape,'C':C.shape,'D':D.shape,'E':E.shape}
print('Legal Combinations:')
for i in combos:
  for j in combos:
    if combos[i][1] == combos[j][0]:
      print(i+j)
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
CD = np.matmul(C,D)
AE = np.matmul(A,E)
BA = np.matmul(B,A)
print('CD:\n',CD,'shape:',CD.shape)
print('\nAE:\n',AE,'shape:',AE.shape)
print('\nBA:\n',BA,'shape: (1,2)') # manual input required because of default treatment of vector-like arrays

print('\n the dimensions of the resulting matrices possess the number of rows of the 1st matrix and the number of columns of the 2nd matrix')
```

## 3.1 Find $IG$ (be sure to show your work) 😃

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```
I = np.eye(2,2)
G = np.array([[12,11],[7,10]])
IG = np.matmul(I,G)
print('IG:\n',IG)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11],[7,10]])
J = np.array([[0,1,2],[7,10,4],[3,2,0]])
print("Determinant of H:",round(np.linalg.det(H)))
print("Determinant of J:",round(np.linalg.det(J)))   # rounding added for aesthetics to combat floating point rounding errors
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
H_inv = np.linalg.inv(H)
J_inv = np.linalg.inv(J)
print('H inverse:\n',H_inv)
print('\nJ inverse:\n',J_inv)
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not?

```
HH_inv = np.matmul(H,H_inv)
J_invJ = np.matmul(J_inv,J)
print('HH^-1\n',HH_inv)
print('\nJ^-1 J\n',J_invJ)

print('\n They cannot necessarily equal each other because they are of different sizes \nhowever they are both identity matrices, because any matrix multiplied by its inverse yieldsthe identity matrix')
```
