## 1. Define the following terms:

- **Neuron:**
Same thing as a Node, it's a function which takes the weighted sum of neuron outputs in the previous layer (it it's in the input layer, then it's just a number)
- **Input Layer:**
The first layer of neurons which are each contain just values of the object being analyzed/predicted, etc.
- **Hidden Layer:**
Layer of neurons each of which recieve a weighted sum signal from the previous layer's neurons and will in turn have weighted sum values which are returned to neurons in the next layer (the values of these weights, while important are often not handled directly by the user and thus are "hidden")
- **Output Layer:**
The final output, number or vector, of a neural network
- **Activation:**
The process of determining significance of neuron's firing, this is done through an activation function which will either serve to drive the signal towards a 1 or 0 (on or off) in the case of sigmoid activation, or amplify it in the case of relu
- **Backpropagation:**
The portion of training responsible for updating the neuron weights, often done with gradient descent or a similar algorithm

### Improve baseline with Randomized search Cross validation

```python
def create_model(fraction=.2):
  model = Sequential()
  model.add(Dense(5, input_dim=13, activation='relu'))
  model.add(Dropout(fraction))
  model.add(Dense(7, activation='relu'))
  model.add(Dropout(fraction))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model
```

```python
model = KerasClassifier(build_fn=create_model, verbose=0)

param_distributions = {'batch_size' : [10, 20, 30, 40, 50, 100],
              'fraction' : [0, .1, .2, .3, .4, .5],
              'epochs': [100, 200]}

search = RandomizedSearchCV(model, param_distributions=param_distributions, verbose=1)
```

```python
search.fit(X, y, verbose=2)
```
