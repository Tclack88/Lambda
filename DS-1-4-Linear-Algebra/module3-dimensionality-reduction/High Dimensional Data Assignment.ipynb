# Vertical Line Test

## 1.1 Create two graphs, one that passes the vertical line test and one that does not.

```
x_vals = np.linspace(-5,5,100)
y_vals = np.sin(x_vals)
plt.plot(x_vals,y_vals)
plt.title('Vertical Line Test');


x_vals2 = np.linspace(-1,1)
y_vals2 = np.sqrt(1-x_vals2**2)
plt.plot(x_vals2,y_vals2,color='C1')
plt.plot(x_vals2,-1*y_vals2,color='C1')
plt.axvline(x=-.5,color='red')
plt.legend(['sine - passes vertical line test','circle - fails vertical line test'], loc='lower right')
plt.axis('equal');
```

## 1.2 Why are graphs that don't pass the vertical line test not considered "functions?"

Functions take one input and give one output (one x to one y). A vertical line passing through two or more points implies there's more than one y-value for each x-value and so is not a function.

## 2.1 Which of the following relations are functions? Why?

\begin{align}
\text{Relation 1: } \{(1, 2), (3, 2), (1, 3)\}

\end{align}

```
rel1 = [(1,2),(3,2),(1,3)]
rel2 = [(1,3),(2,3),(6,7)]
rel3 = [(9,4),(2,1),(9,6)]
rel4 = [(6,2),(8,3),(6,4)]
rel5 = [(2,6),(2,7),(2,4)]

relations = [rel1,rel2,rel3,rel4,rel5]

for i,relation in enumerate(relations):
  firsts = [tup[0] for tup in relation]
  firsts_set = set(firsts)
  if len(firsts) == len(firsts_set):
    print("relation",i+1,'is a function ::: because there are no points which lie together in the same vertical line (and thus are one-to-one)')
```

### Step 1 - separate into X and Y (labels and target)

```
Y = national.VALUE                      # taget
X = national.drop(columns=['VALUE'])    # labels
```

### Step 2 & 3 - Standardize: Center each column at 0, divide by standard deviation

```
X = X - X.mean()
X = X/X.std()
X.head()
```

### Step 4 - Find covariance matrix

```
# couldn't figure out how to do this with mutliplying the transpose by itself
cov = X.cov()
cov
```

### Step 5 & 6 - find eigenvalues and eigenvectors, get the two largest

```
e_vals,e_vecs = np.linalg.eig(cov)

#print('eigenvalues:',e_vals)
# print('\neigenvectors:',e_vecs)

e1,e2 = e_vals[0].real,e_vals[1].real   # return real components since imaginary components are 0
v1,v2 = e_vecs[0].real,e_vecs[1].real   # no imaginary components is a consequence of symmetric matrices
print('Two largest e-values:\n',e1,e2)
print('\nTwo largest e-vectors:\n',v1,v2)
```

### Step 7 - Find principal components and plot

```
X = national.drop(columns=['VALUE']) # bring back the original X to perform the projections from

pc1 = np.dot(X,v1) # v1 and v2 are normalized already so no need to divide by v1 dot v1 for example
pc2 = np.dot(X,v2)

# Unfortunately, I thought we would be projecting this onto v1 and v2 respectively, but the dimensions of this dot product are ~ 64,000
# and so it's not a scalar as I assumed it would be to place as the coefficient in front of v1 and v2
```
